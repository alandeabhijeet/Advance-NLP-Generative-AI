Why Can't You Use Custom Names?
The Hugging Face Seq2SeqTrainer and transformers library expect:

"input_ids" → Model input (tokenized text)
"labels" → Model output (tokenized target)
"attention_mask" → Attention mask (to ignore padding)

```python
def preprocess_function(examples):
    model_inputs = tokenizer(
        examples["document"], max_length=1024, truncation=True
    )

    labels = tokenizer(
        examples["summary"], max_length=182, truncation=True
    )

    if isinstance(labels["input_ids"][0], int):
        labels["input_ids"] = [
            token if token != tokenizer.pad_token_id else -100 for token in labels["input_ids"]
        ]
    else:
        labels["input_ids"] = [
            [(token if token != tokenizer.pad_token_id else -100) for token in label]
            for label in labels["input_ids"]
        ]

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

```

```python
import evaluate

rouge = evaluate.load("rouge")
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    labels = [[token if token != -100 else tokenizer.pad_token_id for token in label] for label in labels]

    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    result = rouge.compute(predictions=decoded_preds, references=decoded_labels)

    if isinstance(result, float) or isinstance(result, np.float64):
        return {"rouge_score": result}

    return {key: value.mid.fmeasure if hasattr(value, "mid") else value for key, value in result.items()}
```

```python
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
```

```python
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=8,
    save_total_limit=2,
    predict_with_generate=True,
    fp16=True,
    num_train_epochs=1
)


trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
trainer.save_model("./saved_model")
tokenizer.save_pretrained("./saved_model")
```
