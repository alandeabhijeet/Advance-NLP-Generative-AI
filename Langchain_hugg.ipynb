# Api using langchian

```python
repo_id = "deepseek-ai/DeepSeek-R1"
```

```python
from langchain import HuggingFaceHub
```

```python
llm = HuggingFaceHub(repo_id=repo_id ,model_kwargs={"temperature":0.1,"max_length":64} )
```

```python
llm.invoke("Hello who are you ?")
```

```python
from langchain import PromptTemplate , LLMChain
```

```python
tem = """
This is User Question : {qq}
for this to the point short answer if Don't sure they tell Sorry
"""
```

```python
promt = PromptTemplate(template= tem , input_variables=["que"])
```

```python
que = "who is first PM of India ?"
```

```python
llchain = LLMChain(llm = llm , prompt = promt)
```

```python
llchain.invoke(que)
```
